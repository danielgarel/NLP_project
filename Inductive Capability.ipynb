### Saving the indices of test set requiring inductive prediction

```python
# load document list
doc_name_list = []
doc_train_list = []
doc_test_list = []
with open('data/' + 'quora_split_balanced_subset' + '.txt', 'r') as f:
    for line in f.readlines():
        doc_name_list.append(line.strip())
        temp = line.split("\t")

        if temp[2].find('test') != -1:
            doc_test_list.append(line.strip())
        elif temp[2].find('train') != -1:
            doc_train_list.append(line.strip())

```

```python

# load raw text
doc_content_list1 = []
doc_content_list2 = []

with open('data/corpus/' + 'question1' + '.clean.txt', 'r') as f:
    for line in f.readlines():
        doc_content_list1.append(line.strip())


with open('data/corpus/' + 'question2' + '.clean.txt', 'r') as f:
    for line in f.readlines():
        doc_content_list2.append(line.strip())
```

```python
# map and shuffle
train_ids = []
for train_name in doc_train_list:
    train_id = doc_name_list.index(train_name)
    train_ids.append(train_id)

test_ids = []
for test_name in doc_test_list:
    test_id = doc_name_list.index(test_name)
    test_ids.append(test_id)

ids = train_ids + test_ids
print(len(ids))

```

```python
# build corpus vocabulary
train_word_set = set()

for i in train_ids:
    words1 = doc_content_list1[i].split()
    words2 = doc_content_list2[i].split()
    train_word_set.update(words1)
    train_word_set.update(words2)

    
vocab = list(train_word_set)
vocab_size = len(vocab)
```

```python
#Adds index if not a strict subset
inductive_ind = []
for i in range(len(test_ids)):
    words1_test = doc_content_list1[test_ids[i]].split()
    words2_test = doc_content_list2[test_ids[i]].split()
    if not set(words1_test + words2_test).issubset(train_word_set):
        inductive_ind.append(i)
```

```python
#Adds index only if more than 15% of the sentence is not a subset
inductive_ind = []
for i in range(len(test_ids)):
    words1_test = doc_content_list1[test_ids[i]].split()
    words2_test = doc_content_list2[test_ids[i]].split()
    if len(set(words1_test + words2_test).intersection(train_word_set))/len(set(words1_test + words2_test))<=0.85:
        inductive_ind.append(i)
```
